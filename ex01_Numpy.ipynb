### 영화 평점 데이터 분석하기

```python
# 데이터 타입 변경
# dtype
data = np.loadtxt('ratings.txt', delimiter = "::",dtype='int')
data
```

```python
# 전체 평점 평균 구하기
np.mean(data[ : , 2])
```

```python
# 각 사용자별 평점 평균 구하기
# 리스트에 [사용자, 사용자의 평점평균]
# 1번 사용자 평점 평균 구하기 > 1 ~ 6040
user_list = []
# data[:,0] 전체 정보에서 id 값만 출력
user = data[data[:,0]==1] # 전체 정보에서 id값이 1인 값만 출력
# 출력된 값(user)에서 평점값만 추출한 후 평균 구하기
user_list.append([1, np.mean(user[ : ,2]) ]) # 1번 사용자에 대한 평점 평균
user_list

i = 1
while i<6040:
    user = data[data[:,0]==i]
    user_list.append([i, np.mean(user[ : ,2])])
    i+=1
user_list
```

```python
user_list = []
for i in range(1,6041): # 1 ~ 6040까지 반복되는 반복문
    user = data[data[:,0]==i] # 반복될 때마다 다른 유저(1 ~ 6040)의 정보를 출력
    user_list.append([i, np.mean(user[ : ,2]) ])
user_list
```

```python
# 각 사용자별 평균 평점이 4점 이상인 사용자 구하기
user_list = []
for i in range(1,6041):
    user = data[data[:,0]==i]
    if np.mean(user[:,2]) >=4:
        user_list.append(i)
user_list
```

```python
# 각 사용자별 평균 평점이 4점 이상인 사용자 구하기
user_np = np.array(user_list)
user_np
```

```python
over4bol = user_np[:,1] >= 4
```

```python
user_np[over4bol,0].astype('int') # 평균 평점이 4점 이상인 사용자의 모든 정보
```
