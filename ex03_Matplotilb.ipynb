```python
# 한 영역 안에 여러 그래프 그리기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, color='r', label='x') # 그래프 이름
plt.plot(y, color='k', label='y')
plt.plot(z, color='m', label='z')
plt.legend(loc = 'best') # 범례 띄우기
plt.show()

#	best
#	upper right
#	upper left
#	lower left
#	lower right
#	right
#	center left
#	center right
#	lower center
#	upper center
#	center
```

```python
plt.plot(x, color='r', label='엑스') # 그래프 이름
plt.plot(y, color='k', label='와이')
plt.plot(z, color='m', label='제트')
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('x,y,z 그래프')
plt.legend(loc = 'best') # 범례 띄우기
plt.show()
```

#### 차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도를 알아보고 시각화

```python
# 1. 차대차 사건 중
data['사고유형_대분류'] == '차대차'
# 2. (1)죽거나 다친 사람(사상자)이 많은 (2) 발생지 시도
# 3. 시각화
```

```python
acc = data[data['사고유형_대분류']=='차대차']
acc
```

```python
acc['발생지시도'].unique()
```

```python
acc['발생지시도'].value_counts()
```

```python
acc['사상자수'].value_counts()
```

```python
acc[['사상자수', '발생지시도']]
```

```python
# 그룹화 - groupby('그룹화 시킬 컬럼').sum() 또는 mean() 등
acc.loc[:,['사상자수', '발생지시도']].groupby('발생지시도').sum()
```

```python
acc_g = acc[['사상자수', '발생지시도']].groupby('발생지시도').sum()
acc_g
```

```python
x = acc_g.index
y = acc_g['사상자수']
```

```python
# 시각화
plt.figure(figsize=(10,5)) # 가로, 세로
plt.bar(x,y)
plt.title('차vs차 교통사고의 사상자 수')
plt.ylabel('사상자 수')
plt.xlabel('지역')
plt.show()
```
