```python
# value를 기준으로 데이터 합치기
df5 = pd.DataFrame({'key': ['K0','K2','K3','K4'],
'A':['A0','A1','A2','A3'],
'B':['B0','B1','B2','B3']})


df6 = pd.DataFrame({'key': ['K0','K1','K2','K3'],
'C':['C0','C1','C2','C3'],
'D':['D0','D1','D2','D3']})
```

```python
# merge
# 데이터를 기준으로 df 합침
# 기준이 같은 것만 출력 : how = inner
# 두개 밖에 못합침
pd.merge(df5, df6, on = 'key')
```

```python
pd.merge(df5, df6, how='outer')
```

```python
# how : left, right (left : 왼쪽 꺼 기준 정렬, right: 오른쪽 꺼 기준 정렬)
pd.merge(df5, df6, how='right')
```
